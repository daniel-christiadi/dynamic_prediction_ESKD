```python
import pandas as pd
from tsfresh import extract_features
from tsfresh.feature_extraction import ComprehensiveFCParameters
```

```python
def feature_extractor_comprehensive(file):
    df = pd.read_csv(file)
    extracted_features = extract_features(df, column_id='id', column_sort='relyear', column_kind='test', column_value='value',
                                          default_fc_parameters=ComprehensiveFCParameters())
    return extracted_features
```

```python
extracted_features = feature_extractor_comprehensive(file="../ML for ESKD/ts_2y.gz")
```

```python
extracted_features.to_csv("../ML for ESKD/ts_2y_feat.gz", index=True, compression='gzip')
```
